```python
# inputing categorical missing data with the mode value

col_name=["Gender","Married","Dependents","Self_Employed"]


for x in col_name:
    loantrain[x].fillna(loantrain[x].mode()[0],inplace=True)
    loantest[x].fillna(loantest[x].mode()[0],inplace=True)
    
```

```python
# inputting numerical missing data with mean value

col_names=['LoanAmount','Loan_Amount_Term']

for x in col_names: 
    loantrain[x].fillna(round(loantrain[x].mean(),0),inplace=True)
    loantest[x].fillna(round(loantest[x].mean(),0),inplace=True)
```

```python
print(loantrain.isnull().sum())
print(loantest.isnull().sum())
```

```python
# inpitting values for credit history columns differently


loantrain["Credit_History"].fillna(value=0,inplace=True)
loantest["Credit_History"].fillna(value=0,inplace=True)
```

```python
print(loantrain.isnull().sum())
print(loantest.isnull().sum())
```

```python
# transfering categorical to numerical


from sklearn.preprocessing import LabelEncoder
colname=['Gender','Married','Education','Self_Employed','Property_Area','Loan_Status','Dependents']

le=LabelEncoder()

for x in colname:
    loantrain[x]=le.fit_transform(loantrain[x])
```

```python
# CREATING AND TESTING TRAINING DATA SETS AND RUNNING THE MODEL
```

```python
x_train=loantrain.values[:,0:-1]
y_train=loantrain.values[:,-1]
y_train=y_train.astype(int)
```

```python
x_train.shape
```

```python
x_test=loantest.values[:,:]
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)


x_test=scaler.transform(x_test)
```

```python
print(x_train)

print()

print(x_test)
```

```python
from sklearn.svm import SVC
svc_model=SVC(kernel='rbf',C=20,gamma=0.01)
svc_model.fit(x_train,y_train)
y_pred=svc_model.predict(x_test)
print(list(y_pred))
```

```python
svc_model.score(x_train,y_train)


# score
#y_pred=SVC_MODEL.PREDICT(X_TRAIN)
#accuracy_score(y_train,y_pred)
```

```python
from sklearn.metrics import confusion_matrix,classification_report

y_pred_new=svc_model.predict(x_train)
print(confusion_matrix(y_train,y_pred_new))
print(classification_report(y_train,y_pred_new))
```
